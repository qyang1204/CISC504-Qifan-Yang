```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataframe1 = pd.read_csv(r'C:\Users\laoyy\Desktop\Certificate_of_Need_Applications__Beginning_1974.csv')
pal = dict(Diagnostic and Treatment Center="blue", Hospital="red", Certified Home Health Agency="yellow", Residential Health Care Facility="green", Licensed Home Care Services Agency="purple", Hospice="pink", Long Term Home Health Care Program="orange")
g = sns.FacetGrid(dataframe1, hue="Facility Type", palette=pal)
g.map(plt.scatter, "Total Project Cost", "Application Received Year", s=100, alpha=.7, linewidth=.8, edgecolor="white")
g.add_legend()
```
