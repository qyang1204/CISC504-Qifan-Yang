```python
import random
i = int(input('Enter number: '))
n = random.randint(1, 100)
while n in range(100):
    if n<i:
        print(str(n)+ ' too low')
        n=n+1
    if n>i:
        print(str(n)+ ' too high')
        n=n-1
    if n==i:
        print(str(n)+ ' right')
        break
```
